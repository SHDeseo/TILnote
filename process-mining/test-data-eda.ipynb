```python
logdata = {}
input_file_name = "data/Production_Data.csv"

with open(input_file_name, 'r') as input_file:
    lines = input_file.readlines()
    columns = [ a.strip() for a in lines[0].split(',')]
    
    for col in columns:
        logdata[col] = []
    
    for line in lines[1:] :
        record = [el.strip() for el in line.split(',')]
        for idx, col in enumerate(columns):
            if col == 'Activity':
                mergedActivity = record[idx]
                try:
                    logdata[col].append(mergedActivity.split(' - ')[0])
                except:
                    logdata[col].append(mergedActivity)
            else:
                logdata[col].append(record[idx])        
            
```
